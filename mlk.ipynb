```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
img = cv2.imread('mlksmile.jpg')
plt.imshow(img) #shows in BGR (Blue,Green,Red) are the defaults great people!
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) #change that pic to RGB

plt.imshow(img)
```

```python
imgTwo = cv2.imread('mlksmile2.jpg')
plt.imshow(imgTwo)
```

```python
verification = DeepFace.verify(img1_path = "mlksmile.jpg", img2_path = "mlksmile2.jpg")
verification
```

```python
# analyzing the face of martin luther king, jr.

analysis = DeepFace.analyze(img_path = "mlksmile.jpg", actions = ["age", "gender", "emotion", "race"])
print(analysis)
```

```python
# verifying that the two images are the same person.
DeepFace.verify("mlksmile.jpg","mlksmile2.jpg") 
```

```python
# As you can see, the two pictures in this cell is "false" because its not the same person.
DeepFace.verify("mlksmile.jpg","ransmile.jpg")
```
